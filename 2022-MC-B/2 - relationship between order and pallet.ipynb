```python
%matplotlib inline
import csv
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colors
from matplotlib.collections import LineCollection

import pandas as pd
```

```python
orders = pd.read_csv('data/orders.csv')
orders
```

```python
pallets = pd.read_csv('data/pallets.csv')
pallets
```

```python
sku2palletID = dict()
for i, row in pallets.iterrows():
    sku_quantity_list, x, y, pallet_id = row
    # print(sku_quantity_list, x, y, pallet_id)
    sku_quantity_list = sku_quantity_list.split(',')  # e.g. ['104151:9', '840211:35', '1297235:1']
    for sq_str in sku_quantity_list:  # sq_str like '104151:9'
        sku, quantity = sq_str.split(':')
        quantity = int(quantity)
        temp = sku2palletID.get(sku, [])
        temp.append((pallet_id, quantity))
        sku2palletID[sku] = temp
len(sku2palletID), sku2palletID['104151']
```

```python
g1 = 0
e1 = 0
for i, row in orders.iterrows():
    order_id, sku, amount = row
    #print(order_id, sku, amount)
    #print('{};{}'.format(order_id, sku2palletID[str(sku)]))
    if len(sku2palletID[str(sku)]) > 1:
        g1 += 1
        print('{};{}'.format(order_id, sku2palletID[str(sku)]))
    else:
        e1 += 1
        if amount > sum([_[1] for _ in sku2palletID[str(sku)]]):
            print(amount, sku2palletID[str(sku)])
            break
g1, e1
```
