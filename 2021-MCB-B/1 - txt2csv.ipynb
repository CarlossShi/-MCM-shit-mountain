```python
G = nx.Graph()
i = 1
for _ in train_df.values:
    #print(_)
    userID = _[0]
    sequence = _[-1]
    #print(user, sequence)
    for c in sequence.split(';'):
        itemID, clicked, duration = c.split(':')
        if int(clicked):
            #print(itemID, clicked, duration)
            G.add_edges_from([('u' + userID, 'i' + itemID)])
            G.nodes['u' + userID]['duration'] = G.nodes['u' + userID].get('duration', 0) + eval(duration)
            G.nodes['i' + itemID]['duration'] = G.nodes['i' + itemID].get('duration', 0) + eval(duration)

    i += 1
    if i>100000:
        break
user_nodes = [_ for _ in G.nodes if _[0] == 'u']
user_max_d = max([G.nodes[_]['duration'] for _ in user_nodes])
item_nodes = [_ for _ in G.nodes if _[0] == 'i']
item_max_d = max([G.nodes[_]['duration'] for _ in item_nodes])
print(user_max_d, item_max_d)

node_size_base = 100
node_size = [None] * len(G.nodes)
for i, _ in enumerate(G.nodes):
    if _[0] == 'u':
        node_size[i] = node_size_base * G.nodes[_]['duration'] / user_max_d
    elif _[0] == 'i':
        node_size[i] = node_size_base * G.nodes[_]['duration'] / item_max_d
    else:
        assert False

edge_max_d = max([G.nodes[u]['duration']+ G.nodes[i]['duration'] for (u, i) in G.edges])
print(edge_max_d)

width_base = 1
width = [None] * len(G.edges)
for k, (u, i) in enumerate(G.edges):
    width[k] = edge_alpha_base * (G.nodes[u]['duration'] + G.nodes[i]['duration']) / edge_max_d
```

```python
alpha = 10/math.sqrt(len(G.nodes))
alpha = alpha if alpha < 1 else 1
plt.rcParams["figure.figsize"] = (12,40)
nx.draw(G, pos=nx.bipartite_layout(G, nodes=user_nodes), node_size=node_size, alpha=alpha, width=width)
```
